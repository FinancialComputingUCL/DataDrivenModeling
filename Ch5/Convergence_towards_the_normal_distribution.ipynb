# Convergence towards the normal distribution

In this code, the “roll_dice” function simulates rolling a given number of dice (“num_dice”) and returns the sum of their outcomes. The “generate_histogram” function performs the specified number of experiments, each time rolling the specified number of dice and recording the sum of their outcomes. It then plots a histogram of these outcomes.

The code calculates the theoretical mean and standard deviation of the sum of the dice using the formulas based on the number of dice being rolled. It generates a range of x values and uses the normal distribution formula to calculate the corresponding y values for the normal distribution curve. The code then plots the histogram and overlays the normal distribution curve on top of it.

By running this code, you can observe how the distribution of the sum of the dice changes as the number of dice increases. With a single die, the distribution is uniform. With two dice, it takes a triangular form. But as the number of dice increases to four, the distribution starts to resemble a bell-shaped normal distribution, as described in the exercise.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt

def roll_dice(num_dice):
    """Simulate rolling 'num_dice' dice and return the sum of their outcomes."""
    return np.sum(np.random.randint(1, 7, size=num_dice))

def generate_histogram(num_dice, num_experiments, subplot_position):
    """Generate a histogram of the frequency of occurrence of the sum of 'num_dice' dice."""
    # Perform the experiments
    outcomes = [roll_dice(num_dice) for _ in range(num_experiments)]

    # Create a subplot
    plt.subplot(subplot_position)

    # Plot the histogram of outcomes
    plt.hist(outcomes, bins=range(num_dice, 6*num_dice+2), density=True, alpha=0.5, color='blue')

    # Calculate the theoretical mean and standard deviation
    theoretical_mean = num_dice * (1 + 6) / 2
    theoretical_std = np.sqrt(num_dice * (6**2 - 1) / 12)

    # Generate a range of x values
    x = np.linspace(theoretical_mean - 3 * theoretical_std, theoretical_mean + 3 * theoretical_std, 100)

    # Calculate the corresponding y values using the normal distribution formula
    y = 1 / (np.sqrt(2 * np.pi * theoretical_std**2)) * np.exp(-(x - theoretical_mean)**2 / (2 * theoretical_std**2))

    # Plot the theoretical normal distribution curve
    plt.plot(x, y, color='red', linewidth=2)

    # Set plot title
    plt.title(f'Number of Dice: {num_dice}')

# Set the number of experiments
num_experiments = 1000

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Generate histograms for different numbers of dice
generate_histogram(1, num_experiments, 221)
generate_histogram(2, num_experiments, 222)
generate_histogram(3, num_experiments, 223)
generate_histogram(4, num_experiments, 224)

# Set common labels
fig.text(0.5, 0, 'Sum of Dice', ha='center', va='center')
fig.text(0, 0.5, 'Probability Density', ha='center', va='center', rotation='vertical')

# Adjust spacing
plt.tight_layout()

# Display the plots
plt.show()
```
